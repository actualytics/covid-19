## Uploading data

We use the data daily published by [Johns Hopkins'CSSE](https://github.com/CSSEGISandData/COVID-19).

```python
import pandas as pd
import chart_studio.plotly as py
import plotly.graph_objects as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
rawConfirmed=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
rawDeaths=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
rawRecovered=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
```

```python
rawConfirmed
```

Let's reshape the dataframes

```python
countriesCoord=rawConfirmed[['Country/Region','Lat','Long']].drop_duplicates().copy()
countriesCoord
```

```python
confirmed1=rawConfirmed.drop(['Lat','Long','Province/State'],axis=1)
confirmed1=confirmed1.groupby('Country/Region').sum()
confirmed1=confirmed1.T
confirmed1.index.name='date'

deaths1=rawDeaths.drop(['Lat','Long','Province/State'],axis=1)
deaths1=deaths1.groupby('Country/Region').sum()
deaths1=deaths1.T
deaths1.index.name='date'

recovered1=rawRecovered.drop(['Lat','Long','Province/State'],axis=1)
recovered1=recovered1.groupby('Country/Region').sum()
recovered1=recovered1.T
recovered1.index.name='date'
```

add total row

```python
total=confirmed1.T.sum()
confirmed1['Total']=total

total=deaths1.T.sum()
deaths1['Total']=total

total=recovered1.T.sum()
recovered1['Total']=total
```

```python
underTreatment=confirmed1-recovered1-deaths1
confirmed1_shift=confirmed1.shift(1).fillna(0).copy()
```

```python
newCases1=confirmed1-confirmed1_shift
```

```python
def stack_bar(df1,df2,df3,df4,country):
    data=[]
    data.append(go.Bar(x=df1.index,y=df1[country],name='deaths (cumulative)'))
    data.append(go.Bar(x=df2.index,y=df2[country],name='underTreatment (daily)'))
    data.append(go.Bar(x=df3.index,y=df3[country],name='recovered (cumulative)'))
#    data.append(go.Scatter(
#        x=df4.index,
#        y=df4[country],
#        name='new cases (daily)',
#        mode = 'lines+markers',
#        line_shape='spline',
#        marker=dict(
#            size=5,
#            cmid=0,
#            color="black"
#        )
#    ))

    layout=dict(
        barmode='stack',
        yaxis={'categoryorder':'array', 'categoryarray':['deaths (cumulative)','underTreatment (daily)','recovered (cumulative)']}
    )

    fig={'data':data,
        'layout':layout}
    return(fig)

fig=stack_bar(deaths1,underTreatment,recovered1,newCases1,'Total')
iplot(fig)
```

```python
fig=stack_bar(deaths1,underTreatment,recovered1,newCases1,'China')
iplot(fig)
```

### Confirmed cases

```python
def compare_countries(df,countries,timeLimit):
    data=[]
    for country in countries:
        if country!='Total':
            if timeLimit=="absolute":
                data.append(go.Scatter(x=df.index,y=df[country],name=country))
            else:
                time=df[df[country]>timeLimit].index
                data.append(go.Scatter(y=df[df[country]>timeLimit][country],name=country))          

    layout=dict(
        updatemenus=[
            dict(
                buttons=list([
                    dict(
                        args=["yaxis.type", 'log'],
                        label="log",
                        method="relayout"
                    ),
                    dict(
                        args=["yaxis.type", 'linear'],
                        label="linear",
                        method="relayout"
                    )
                ]),
                direction="right",
                pad={"r": 10, "t": 10},
                showactive=True,
                x=0.1,
                xanchor="left",
                y=1.15,
                yanchor="top"
            ),
        ],
        yaxis={
            'type':'log'
        },
        xaxis={
            'title':'day since 10 cases'
        }
    )

    fig={'data':data,
        'layout':layout}
    
    return(fig)

fig=compare_countries(confirmed1,['China','France','Italy','Germany','Iran','Italy','US','United Kingdom','Canada','Korea, South'],10)
iplot(fig)
```

### Deaths

```python
fig=compare_countries(deaths1,['China','France','Italy','Germany','Iran','Italy','US','United Kingdom','Canada','Korea, South'],10)
iplot(fig)
```
